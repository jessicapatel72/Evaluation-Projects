### Remove Outliers

As we see outliers present in some columns. So, first need to remove it.

### 1. Outlier removal using Zscore


```python
#In Zscore technique taking standard deviation 3
#for Zscore outlier removal technique import library from scipy

from scipy.stats import zscore

z_score= zscore(df_baseball[[ 'R', '2B', '3B','CG','AB', 'HR', 'BB', 'SO', 'SB',
                             'RA', 'ER', 'ERA', 'SHO', 'SV', 'E']]) #Put continuous/float input variable in z-score 

abs_z_score = np.abs(z_score)

data = (abs_z_score < 3).all(axis = 1)

new_df_baseball = df_baseball[data]

print("shape before :", df_baseball.shape)
print("shape after :", new_df_baseball.shape)
print("Percentage Loss :", (df_baseball.shape[0]-new_df_baseball.shape[0])/df_baseball.shape[0])
```

3.333% data loss after using zscore technique

96.667% data remains

## 2. Outlier Removing using IQR

### from boxplot in EDA, we came to know that outliers present in following columns.

### Visualize data again to check outliers are present at low side or high side


```python
data = df_baseball
```

```python
#Relationship Visualizing

plt.figure(figsize = (15,45))
plot_number = 1

for i in df_baseball:
    if plot_number <= 18: # as we see there are 17 columns in the data
        plt.subplot(9,2,plot_number)
        sns.boxplot(df_baseball[i], color = 'c')
        plt.xlabel(i,fontsize=14)
        
    plot_number += 1
plt.show()
```

As shown in plot outliers are present in R, ERA, SHO, SV and E columns, so remove it first.

```python
#find the IQR (Inter Quantile Range) to identify outliers
#formula for finding IQR

#1st quantile 25%
q1 = data.quantile(0.25)

#3rd quantile 75%
q3 = data.quantile(0.75)

#IQR = Inter Quantile Range
iqr = q3-q1
```

```python
data.describe()
```

### Outlier detection formula

- Higher side ==> Q3 + (1.5 * IQR)
- Lower side ==> Q1 - (1.5 * IQR)

as outliers present in higher side only so we will go with higher side formula

```python
#as shown in boxplot we have outlier in higher side of R
#Remove outliers from higher side so, use higher side formula

R_high = (q3.R + (1.5*(iqr.R)))
R_high
```

```python
index_out = np.where(data['R'] > R_high)
data = data.drop(data.index[index_out])
data.shape
data.reset_index()
```

after removing outliers from R column, 29 rows will remain in dataframe.

```python
#as shown in boxplot we have outlier in higher side of ERA
#Remove outliers from higher side so, use higher side formula

ERA_high = (q3.ERA + (1.5*(iqr.ERA)))
ERA_high
```

```python
index_out = np.where(data['ERA'] > ERA_high)
data = data.drop(data.index[index_out])
data.shape
data.reset_index()
```

after removing outliers from ERA column, 28 rows will remain in dataframe.

```python
#as shown in boxplot we have outlier in higher side of SHO
#Remove outliers from higher side so, use higher side formula

SHO_high = (q3.SHO + (1.5*(iqr.SHO)))
SHO_high
```

```python
index_out = np.where(data['SHO'] > SHO_high)
data = data.drop(data.index[index_out])
data.shape
data.reset_index()
```

After removing outliers from SHO column 26 rows will remain in dataframe.

```python
#as shown in boxplot we have outlier in higher side of SV
#Remove outliers from higher side so, use higher side formula

SV_high = (q3.SV + (1.5*(iqr.SV)))
SV_high
```

```python
index_out = np.where(data['SV'] > SV_high)
data = data.drop(data.index[index_out])
data.shape
data.reset_index()
```

After removing outliers in SV column, 25 rows will remain in dataframe.

```python
#as shown in boxplot we have outlier in higher side of E
#Remove outliers from higher side so, use higher side formula

E_high = (q3.E + (1.5*(iqr.E)))
E_high
```

```python
index_out = np.where(data['E'] > E_high)
data = data.drop(data.index[index_out])
data.shape
data.reset_index()
```

After removing outliers in E column, 20 rows will remain in dataframe.

## Finding best random state


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
maxAccuracy = 0
maxRandomState = 0
for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.30, random_state=i)
    dtr_= DecisionTreeRegressor()
    dtr_.fit(x_train, y_train)
    pred = dtr_.predict(x_test)
    acc=r2_score(y_test, pred)
    if acc>maxAccuracy:
        maxAccuracy=acc
        maxRandomState=i
print("Best accuracy is ",maxAccuracy, "on Random_state ", maxRandomState)
```
